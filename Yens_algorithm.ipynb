## Yen's Algorithm
K Shortest Path Problemを求めるアルゴリズム

```python
import sys
import math
from heapq import *
from collections import defaultdict
import random
import networkx as nx
import matplotlib.pyplot as plt
from typing import List
sys.setrecursionlimit(10**7)
```

---
### グラフの作成

```python
edges = [[] for _ in range(6)]

edges[0].append((3, 1))
edges[0].append((2, 3))
edges[1].append((4, 2))
edges[2].append((2, 4))
edges[2].append((1, 5))
edges[3].append((1, 1))
edges[3].append((2, 2))
edges[3].append((3, 4))
edges[4].append((2, 5))
```

```python
# 有向グラフを作成
G = nx.DiGraph()

# エッジを追加（重み付き）
for i, adj in enumerate(edges):
    for length, to in adj:
        G.add_edge(i, to, weight=length)

# レイアウト（見やすく配置）
pos = nx.spring_layout(G, seed=42)

# ノードとエッジの描画
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=1000, arrowsize=20)

# エッジラベル（重み）を表示
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

plt.title("Graph Visualization (edge weights = lengths)")
plt.show()
```

```python
# G = nx.DiGraph()
G = nx.Graph()
G.add_weighted_edges_from([
    ("A", "B", 1),
    ("B", "C", 2),
    ("A", "C", 2),
    ("C", "D", 1),
    ("B", "D", 4)
])

paths = nx.shortest_simple_paths(G, "A", "D", weight="weight")

for i, path in zip(range(4), paths):
    print(f"{i+1}番目: {path}, cost = {nx.path_weight(G, path, 'weight')}")
```
